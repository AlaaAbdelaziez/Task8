### Arab Kaggle Users by Programming Language Distribution

```python
lang_freq=pd.Series([])
Q7=df.columns[df.columns.str.contains('Q7')]
Q7
```

```python
for q in Q7:
    lang_freq=lang_freq.append(df3[q].value_counts())
    
lang_freq=lang_freq.sort_values(ascending=False)
```

```python
plt.pie(lang_freq,explode=explode,labels= lang_freq.index,startangle=10,autopct="%1.3f%%",rotatelabels=True)
plt.rc('xtick', labelsize=15)

plt.show()
```

```python
fig , axes = plt.subplots(nrows=1, ncols=2,figsize=(30,15))
fig.patch.set_facecolor('#F2F2F2')

explode=[0.05 for i in range(len(Q7))]
color=plt.get_cmap('cool')(np.linspace(.1, .5, len(lang_freq)))

axes[1].pie(lang_freq,explode=explode,labels= lang_freq.index,colors=color,startangle=10,autopct="%1.1f%%",rotatelabels=True)


axes[0].bar(lang_freq.index,lang_freq,color=color)
axes[0].set_xlabel('Programming Language', weight ='semibold',fontname='monospace',size=30)
axes[0].set_ylabel('Count', weight ='semibold',fontname='monospace',size=30)
axes[0].tick_params(labelrotation=80,labelsize=25)
axes[0].patch.set_facecolor('#F2F2F2')
axes[0].spines['bottom'].set_color('black')
axes[0].spines['left'].set_color('black')
axes[0].spines['top'].set_visible(False)
axes[0].spines['right'].set_visible(False)

axes[0].grid(color=color[3],alpha=0.5, linestyle='--')


plt.suptitle('Programming Language Distribution of Kaggle Users', weight ='bold',fontname='monospace', size=60)


plt.show()
```
