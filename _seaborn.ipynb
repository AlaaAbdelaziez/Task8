```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
flights = sns.load_dataset('flights')
flights
```

```python
feb_flights = flights[flights.month=="Feb"]
feb_flights
```

```python
sns.set_style("darkgrid")
sns.lineplot(x='year',y='passengers',data=feb_flights)
plt.xlabel("Year")
plt.title("Year vs Passengers")
plt.show()
```

```python
sns.lineplot(x='year',y='passengers',data=flights,hue="month")
```

```python
dash_styles = ["",
               (4, 1.5),
               (1, 1),
               (3, 1, 1.5, 1),
               (5, 1, 1, 1),
               (5, 1, 2, 1, 2, 1),
               (2, 2, 3, 1.5),
               (1, 2.5, 3, 1.2),
               (2,2),
               (1,2,1,1),
               (1,1.5),
               (1,3,1,3)]
sns.lineplot(data=flights, x="year", y="passengers", style="month",dashes=dash_styles)
```

```python
plt.figure(figsize=(10,8))
sns.lineplot(x='year',y='passengers',data=flights,size="month",hue="month")
```
