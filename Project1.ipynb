```python
df['education']=df['education'].apply(lambda x:'others' if(x==5 or x==6 or x==0)else x)
df['marriage']=df['marriage'].apply(lambda  x: x if(x=='single'or x=='married' ) else 'others')
```

```python
df.describe()
```

```python
def remove_outlier(df):
    Q1=df.quantile(0.25)
    Q3=df.quantile(0.75)
    iqr=Q3-Q1
    df[((df<Q1-1.5*iqr)|(df>Q3+1.5*iqr))]=np.nan
    df.dropna()
    return df
df=remove_outlier(df)    
 
    
```
